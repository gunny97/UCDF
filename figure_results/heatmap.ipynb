```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = [10, 8]

ablation_df = pd.read_excel('Results_analysis.xlsx', sheet_name='Ablation')
```

```python
ablation_df.set_index("subspace_size", inplace=True)
```

```python
ax = sns.heatmap(ablation_df, 
           vmin=50,
           vmax=100,
           cbar=True,
           annot=True, annot_kws={'size':10},
           cmap = sns.cm.rocket_r
           )


ax.set_xlabel('the number of queries',size=20)
ax.set_ylabel('subspace size',size=20)

# plt.title('AGNews Accuracy', fontsize=30)

plt.savefig('agnews_ablation.pdf', dpi=300)

plt.show() 
```

### Bar Plot

```python
ablation_df = pd.read_excel('Results_analysis.xlsx', sheet_name='augmentation')
ablation_df.drop('Full-data',axis=1,inplace=True)
ablation_df = ablation_df[['UCDF','Back Translation', 'EDA', 'w/o DA']]
ablation_df = ablation_df.T
```

```python
ablation_df.reset_index(drop=False, inplace=True)
ablation_df.columns = ["Data Augmentation Technique", 'Accuracy']
```

```python
bar = sns.barplot(
    data = ablation_df,
    x= "Data Augmentation Technique",
    y= "Accuracy"
)

# bar.axhline(78.95,color='r')
# bar.axhline(88.09,color='r', linestyle='--')

bar.bar_label(bar.containers[0])

plt.ylim(70, 90)

plt.savefig('augmentation_technique.pdf', dpi=300)

plt.show()
```
